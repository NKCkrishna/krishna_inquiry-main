```python
%%writefile src/App.js
import React, { useState } from 'react';

function App() {
  const [role, setRole] = useState('user');

  return (
    <div>
      <h1>Real-Time Inquiry System</h1>
      <button onClick={() => setRole('user')}>User View</button>
      <button onClick={() => setRole('admin')}>Admin View</button>

      {role === 'user' ? <UserInterface /> : <AdminInterface />}
    </div>
  );
}

function UserInterface() {
  return <div>User Interface: Submit and track inquiries.</div>;
}

function AdminInterface() {
  return <div>Admin Interface: Manage and respond to inquiries.</div>;
}

export default App;
```
